```python
data_AB = data['failures']

#quartiles
data_AB_q1 = data_AB.quantile(0.25)
data_AB_q2 = data_AB.quantile(0.5)
data_AB_q3 = data_AB.quantile(0.75)

#inter quartile range
data_AB_iqr = data_AB_q3 - data_AB_q1

#lower whisker point
data_AB_lower = data_AB_q1 - (1.5 * data_AB_iqr)

#upper whisker point
data_AB_upper = data_AB_q3 + (1.5 * data_AB_iqr)

data_AB_outliers = data_AB[(data_AB < data_AB_lower) | (data_AB > data_AB_upper)]
data_AB_outliers
```
